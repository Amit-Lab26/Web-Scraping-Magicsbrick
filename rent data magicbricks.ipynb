```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url="https://www.magicbricks.com/independent-house-for-rent-in-thane-pppfr"
```

```python
page=requests.get(url)
```

```python
print(page)
```

```python
soup=BeautifulSoup(page.text,'html.parser')
print(soup)
```

```python
lists=soup.find_all('div',attrs={'class':'mb-srp__list'})
print(lists)
```

```python
lists=soup.find_all('div',attrs={'class':'mb-srp__list'})
name1=[]
rent1=[]
location1=[]
for a in lists:
    name=a.find('h2',attrs={'class':'mb-srp__card--title'})
    rent=a.find('div',attrs={'class':'mb-srp__card__price--amount'})
    location=a.find('div',attrs={'class':'mb-srp__card--desc--text'})
    name1.append(name.text)
    rent1.append(rent.text)
    location1.append(location.text)
    
Data_Bang_Frame = pd.DataFrame({"Name":name1,"Rent":rent1,"Location":location1})
#df.to_csv(‘Data.csv’)
```

```python
Data_Bang_Frame.to_csv('rentData.csv')
```
